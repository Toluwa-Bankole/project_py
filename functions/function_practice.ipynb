```python
# function to derive total number of seconds in a given time
def total_seconds(hours, minutes, seconds):
    total_seconds  = (hours * 3600) + (minutes * 60) + seconds
    return total_seconds
```

```python
Seconds = total_seconds(2, 30, 45)
Seconds
```

```python
# Define a function named `total_sales`
def total_sales(quantity,sales):
    total_sales = quantity * sales
    # Return a value
    return total_sales
```

```python
# Define a function named `total_sales`
# Numbers have to come before the parameters i.e 7.99*quantity and not quantity*7.99
def total_sales_1(quantity):
    sales = 7.99*quantity
    # Return a value
    return sales
```

```python
# Test the functions out
total_sales_1(20)
```

```python
def seed_calculator(fountain_side, grass_width):
    # Area of fountain
    fountain_area = fountain_side**2
    # Total area
    total_side = (fountain_side + 2 * grass_width)**2
    # Area of grass boarder
    grass_area = total_side - fountain_area
    # Amount of seed needed (35 g/sq.m)
    seed = grass_area*35
    # Convert to Kg
    seed_per_kg = seed/1000

    return seed_per_kg
```

```python
seed_calculator(12,2)
```
