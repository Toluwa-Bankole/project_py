String Indexing

index() method - Is a string method that outputs the index number of a character in a string.

We can call a method by following the variable with a dot e.g. variable_name.index()

The index() method will return the first position that matches only

```python
name = 'Tolu James'
name.index('s') # to find index number of 's' in the variable name
```

```python
name[2] # to find the character at index 2 in the variable name
```

```python
name[-1] # to find the last character in the variable name
```

```python
name[-2] # to find the second to last character in the variable name
```

String Slicing

Also know as substring, is a method to extract a specific length of character(more than 1 character) from a string using their index number

```python
name[:4] # to find the characters from index 0 to index 3 in the variable name
```

```python
name[1:4] # to find the characters from index 1 to index 3 in the variable name
```

```python
name[5:] # to find the characters from index 5 to the end of the variable name
```

Format() method(i.e. called with the dot notation)

formats and inserts specific substrings into designated places

```python
company = 'DFI'
number = 24647534.585959

print(f"This is {company}.\nWith a networth of ${number:.2f} tech institute in Nigeria.") # .2f means round to 2 decimal places
                                                        # if i used % instead 2f it make multiple the number by 100 percent
                                                                # \n is line breaker i.e. move whatever follows to the next line.
```

```python
company = 'DFI'
number = 24647534.585959

print(f"This is {company}.\nWith a networth of ${number:.2%} tech institute in Nigeria.") # .2f means round to 2 decimal places
                     # if I used 2% instead 2f it make multiple the number by 100 percent and round it up to 2 decimal places
                    # \n is line breaker i.e. move whatever follows to the next line.
```

```python
company = 'DFI'
number = 24647534.585959

print(f"This is {company.title()}.\nWith a networth of ${number:.2f} tech institute in Nigeria.") # Any inbuilt function can be used in f-strings
```

```python
company = 'DFI'
number = 24647534.585959

print(f"This is {company.lower()}.\nWith a networth of ${number:.2f} tech institute in Nigeria.") # Any inbuilt function can be used in f-strings
```

```python
company = 'DFI'
number = 24647534.585959

print(f"This is {company.upper()}.\nWith a networth of ${number:.2f} tech institute in Nigeria.") # Any inbuilt function can be used in f-strings
```

```python
company = 'DFI'
number = 24647534.585959

print(f"This is {len(company)}.\nWith a networth of ${number:.2f} tech institute in Nigeria.") # Any inbuilt function can be used in f-strings
```

Removing whitespaces in strings

```python
company = 'DFI  Nigeria  '
number = 24647534.585959

# this will remove any whitespace from the beginning and end of the string
print(f"This is {company.strip()}.\nWith a networth of ${number:.2f} tech institute in Nigeria.") # Use this if string is a single word

# this will remove any whitespace from the beginning of the string
print(f"This is {company.lstrip()}.\nWith a networth of ${number:.2f} tech institute in Nigeria.")
# this will remove any whitespace from the end of the string
print(f"This is {company.rstrip()}.\nWith a networth of ${number:.2f} tech institute in Nigeria.")
# this will remove any whitespace from the string and join the words with a single space in between
print(f"This is {" ".join(company.split())}.\nWith a networth of ${number:.2f} tech institute in Nigeria.") # Use this if string is not single word
```

Validating strings

To check if a string exist or not

```python
company = 'DFI  Nigeria  '

print(f'This is {company.find('FI')}.') # This give the index number of the first substring
```

```python
company = 'DFI  Nigeria  '

print(f'This is {'F' in company}.') # This a true or false response.
```

Replacing strings

```python
company = 'DFI  Nigeria  '

print(f'This is {company.replace('DFI', 'good')}')
```

Replacing strings and removing whitespaces


```python
company = 'DFI  Nigeria  '

print(f'This is {" ".join(company.replace('DFI', 'good').split())}') # logical order, what you want to replace must be replaced before splitting string
```
