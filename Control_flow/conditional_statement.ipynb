```python
def rescind_access(position):
    #use elif there are more than two conditions you are checking on
    if position == "contract_staff": # condition 1
        print("Access not granted")
    elif position == "Probation": # condition 2
        print("Access denied")
    elif position == "Manager": # condition 3
        print("Access Granted")
    else: # If condition 1 to 3 is not met then end with this
        print("Alert")
```

```python
rescind_access('Terminated')
```

```python
def send_email (num_visit, visits_email,visits_coupon):
    if num_visit>=visits_coupon:
        print("Send email with coupon")
    elif num_visit>=visits_email:
        print('Send email only')
    else:
        print("Not enough visits")
```

```python
send_email(3,5,8)
send_email(5,5,8)
send_email(8,5,8)
```
