Control-flow

While loop -- Keep doing this while something is true. It will repeat a block of code as long as a condition remains true.
Infinite occur when condition never become false
Use when you don’t know how many times the loop will run
Use when you want to repeat an action until the boolean condition changes

```python
x = 1 
 # a counter that controls the logic of the program also called variable initialization (starting 
 # point)


while x <= 7:
#condition to run on is x < 7, so as long as the condition is true, it will keep looping

    print(f"{x} is right") 
    #this is the body of the loop, it will run every time the condition is true

    x=x+2
    # this can also be written as x +=2. It is the increment that changes the value of x, so that 
    # eventually the condition will become false and the loop will stop


print(x) 
# Exit point (loop stops when condition becomes False). This will print the final value of x after the loop has finished.
```

```python
x = 7 
# a counter that controls the logic of the program also called variable initialization (starting point)
number_of_guesses = 0

while number_of_guesses <= 7: 
    # condition to run on is x < 7, so as long as the condition is true, it will keep looping
    # always place your increment immediately after the body of the loop, so that it will run 
    # every time the condition is true.
    number_of_guesses += 1 
    # this can also be written as number_of_guesses = number_of_guesses + 1. It is the increment 
    # that changes the value of number_of_guesses, so that eventually the condition will become 
    # false and the loop will stop

    if number_of_guesses == 3: 
        continue 
    # this will skip the rest of the code in the loop and go back to the beginning of the loop 
    # if the number of guesses is equal to 3  
    elif number_of_guesses == 7: 
        break 
    # this will break out of the loop if the number of guesses is equal to x
    else:
        print(f"{number_of_guesses} is not right") 
        # this will print the number of guesses if it is not equal to x

#print(x) # Exit point (loop stops when condition becomes False). This will print the final value of x after the loop has finished.
```

```python
from time import sleep

n = 3
while n >= 0:
    print(n)
    sleep(1)    # Execution pauses for 1 second
    n = n - 1
```

```python
mins = 10

while mins >= 0:
    if mins == 5:
        print(f"Place your reservation soon! {mins} minutes remaining.")
    elif mins ==2:
        print(f"Don't lose your seat. {mins} minutes remaining.")
    elif mins == 0:
        print(f"User timed out.")
    else:
        print(mins)
    mins = mins - 1 # Aternatively : mins -= mins
```
