Naming styles - To make variable, function and file names easier to read.
* Camel case - Words are written together without spaces, and each word (except sometimes the first) starts with a capital letter e.g. firstName
* Pascal Case (Upper Camel Case) - Very similar to camel case, but every word starts with a capital letter (including the first) e.g. FirstName
* Snake case - Words are lowercase and separated by hyphens (-) e.g. first-name

Escape character - A character that changes the typical behaviour of the characters that follows it. It is a backslash(\) followed by a character you want to insert into a string

* \n - means break into a new line e.g. love = "God is love, \nKnow this"
* \' - means insert a single quotation(')
* \" - means insert a double quotation(")
* \\ - means insert a backslash(\)
* \t - means add a horizonta whitespace i.e. indentation

len() = “How many items are inside this thing?”

* In a string → items = characters e.g len("Hello")  result: 5
* In a list → items = elements e.g len((1,2,3,4))    result: 4
* In a tuple → items = elements e.g len([1,2,3])     result: 3
* In a dict → items = key-value pairs e.g len({"a":1, "b":5})  result: 2

Control-flow

* While loop -- Keep doing this while something is true. It will repeat a block of code as long as a condition remains true.
Infinite occur when condition never become false
Use when you don’t know how many times the loop will run

* for loop -- for loop is used to iterate over a sequence (list, string, range, etc.)
When you know the number of iterations
When looping through a collection of items
It has a keyword 'in'

for variable_name in range():
    loop body

for variable_name in iterable:
    loop body

An iterable is anything you can loop over, such as:
* list
* string
* tuple
* dictionary
* range()

WORKING WITH DATAFRAME

* Data Cleaning - Regex approach

        - Removing whitespaces 

        Examples: df["columnA"] =  df["columnA"].replace(r"\s+"," ", regex=True)

        \s+ means match one or more whitespace character

        - Remove numbers from text 

        Examples: df["columnA"] =  df["columnA"].replace(r"\d+","", regex=True)

        - Remove text, underscore from value 

        Examples: df["columnA"] =  df["columnA"].replace(r"\w+","", regex=True)

        - Remove special characters like &, #, @ etc.

        Examples: df["columnA"] =  df["columnA"].replace(r"[^A-Za-z0-9 ]","", regex=True)

        - Replace multiple spelling errors from a column

        Examples: df["columnA"] =  df["columnA"].replace(r"old_text1 | old_text2 | old_text3","new_text", regex=True)

        - Replace a specific character

        Examples: df["columnA"] =  df["columnA"].replace(r"-","new_character", regex=True)

STRING SLICING

Use to extract a range of character in a string

```python
name[:4] # to find the characters from index 0 to index 3 in the variable name
```

```python
name[1:4] # to find the characters from index 1 to index 3 in the variable name
```

```python
name[5:] # to find the characters from index 5 to the end of the variable name
```
